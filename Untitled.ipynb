```python
# Function to calculate score
bobot = {
    "hadir": 10,
    "tugas": 20,
    "uts": 30,
    "uas": 40
}

def hitung_nilai(sks, hadir, tugas, uts, uas, bobot=bobot):
    max_hadir = 0
    # Determine maximum attendance based on SKS
    if sks == 2:
        max_hadir = 14
    elif sks == 3:
        max_hadir = 21
    else:
        return "SKS tidak valid"
    
    # Calculate weighted scores
    # Attendance
    hadir_score = (hadir / max_hadir) * bobot["hadir"]
    
    # Tugas, UTS, UAS
    tugas_score = (tugas / 100) * bobot["tugas"]
    uts_score = (uts / 100) * bobot["uts"]
    uas_score = (uas / 100) * bobot["uas"]
    
    # Total score
    total_score = hadir_score + tugas_score + uts_score + uas_score
    
    return round(total_score, 2)
```

```python
# Function to determine letter grade based on numerical score
def get_grade(nilai):
    if nilai >= 85:
        return "A"
    elif nilai >= 70:
        return "B"
    elif nilai >= 55:
        return "C"
    elif nilai >= 40:
        return "D"
    else:
        return "E"
```

```python
# Creating the results list
results = []

# Iterate over each student
for student in students:
    student_summary = {
        "nim": student["nim"],
        "nama": student["nama"],
        "rangkuman_nilai": []
    }
    
    # Iterate over each subject for the student
    for mk in student["hasil"]:
        # Calculate the final score using the hitung_nilai function
        final_score = hitung_nilai(mk["sks"], mk["hadir"], mk["nilai"]["tugas"], mk["nilai"]["uts"], mk["nilai"]["uas"])
        
        # Check if the final score is valid
        if isinstance(final_score, (int, float)):  # Check if it's a number
            # Create a summary for the course
            course_summary = {
                "mata_kuliah": mk["mata_kuliah"],
                "nilai": f"{final_score} : 0 - 100",
                "Grade": f": {get_grade(final_score)}"
            }
            
            # Add the course summary to the student's summary
            student_summary["rangkuman_nilai"].append(course_summary)
        else:
            # Handle the invalid score case
            course_summary = {
                "mata_kuliah": mk["mata_kuliah"],
                "nilai": "Invalid score",
                "Grade": "N/A"
            }
            student_summary["rangkuman_nilai"].append(course_summary)
    
    # Add the student summary to the results
    results.append(student_summary)

# Print the results
import pprint
pprint.pprint(results)
```
